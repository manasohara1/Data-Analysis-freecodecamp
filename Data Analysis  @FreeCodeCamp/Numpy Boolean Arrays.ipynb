Boolean Arrays

```python
import numpy as np
import sys
a=np.arange(4)
```

```python
a
```

```python
a[0],a[-1]
```

```python
a[[True,False,False,True]]
```

```python
a>=2
```

```python
a[a>=2]
```

```python
a.mean()
```

```python
a[a>=a.mean()]
```

```python
a[~(a>=a.mean())]
#negation
```

```python
a[(a==0) | (a==1)]
```

```python
a[(a>1) & (a<=4)]
```
